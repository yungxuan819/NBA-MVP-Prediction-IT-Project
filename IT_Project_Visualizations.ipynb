```python
stats.corr(numeric_only = True)["Share"].plot.bar()
```

```python
correlation_matrix = stats.corr(numeric_only=True)

# Extract correlation coefficients related to 'Share' column
share_correlation = correlation_matrix['Share']

# Reshape the series for a square box heatmap
reshaped_correlation = share_correlation.values.reshape(1, -1)

# Create a heatmap
plt.figure(figsize=(32, 10))
sns.heatmap(reshaped_correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0, square=True,
            xticklabels=correlation_matrix.columns, yticklabels=['Share'])
plt.xlabel("Player Stats")
plt.title('Correlation Heatmap with Share')
plt.show()
```

```python
#correlations of each features
corr_matrix = stats.corr(numeric_only = True)

#plot heat map
mask = np.zeros_like(corr_matrix)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(15, 15))
    ax = sns.heatmap(corr_matrix, mask=mask, vmax=.3, square=True,cmap="RdYlGn")
```
